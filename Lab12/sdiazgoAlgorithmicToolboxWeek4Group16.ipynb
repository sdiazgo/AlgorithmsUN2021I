##Binary search

```python
def bin_s(x, y, r):
    l = 0

    while l<=r: 
        m = (l+r)//2
        if y > x[m]:
            l = m + 1
        elif y < x[m]:
            r = m - 1
        else:
            return m

    return -1

x = list(map(int, input().split()))
b = list(map(int, input().split()))

n = x[0]
x = x[1:]
y = True

lis = []
xk = b[1:]

for i in xk:
    lis.append(bin_s(x, i, n-1))

print(' '.join([str(i) for i in lis]))
```

##Improving quick sort

```python
import random

def qu_int(a, l, r):
    x = a[l]
    b1 = l
    b2 = l
    for i in range(l + 1, r + 1):
    	if a[i] < x:
        	a[i],a[b1] = a[b1],a[i]
        	b1+=1
        	b2+=1
        	a[i],a[b2] = a[b2],a[i]
    	elif a[i] == x:
        	b2+=1
        	a[i],a[b2] = a[b2],a[i]
    return b1,b2

def quick_sort2(a, l, r):
    if l >= r:
        return
    k = random.randint(l, r)
    a[l], a[k] = a[k], a[l]
    j,k = qu_int(a, l, r)
    quick_sort2(a, l, j-1)
    quick_sort2(a, k+1, r)

n = int(input())

a = list(map(int, input().split()))

quick_sort2(a, 0, n - 1)

for x in a:
    print(x, end=' ')
```

##Organizing a lottery

```python
lista_t = list()
x, y = [int(i) for i in input().split()]

for i in range(x):
    a, b = [int(i) for i in input().split()]
    lista_t.append((a,'l'))
    lista_t.append((b,'r'))

points = input().split()
for i in points:
    lista_t.append((int(i),'p'))

lista_t.sort()

sc = 0
sm = dict()
for i in lista_t:
    if i[1] == 'l': sc += 1
    elif i[1] == 'r': sc -= 1
    else:
        sm[i[0]] = sc

tmp = ''
for i in points:
    tmp += str(sm[int(i)]) + ' '
print(tmp[:-1])
```
